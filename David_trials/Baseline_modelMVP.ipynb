```python
import torch
import torch.nn as nn
import torchvision.models as models
import torch.optim as optim

# The '!' tells the notebook to run a terminal command
!pip install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu121
# Import our custom helper functions
from Helper_funcktions import get_cifar10_dataloaders, train_model, plot_loss_curves

# 1. Setup device (GPU if available, else CPU)
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Training on device: {device}")

# 2. Load the pre-trained ResNet-18 model
baseline_model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)

# 3. Freeze the backbone (so we only train the final layer)
for param in baseline_model.parameters():
    param.requires_grad = False

# 4. Replace the final layer with a STANDARD linear layer (for 10 CIFAR-10 classes)
num_ftrs = baseline_model.fc.in_features
num_classes = 10
baseline_model.fc = nn.Linear(num_ftrs, num_classes)

print("Baseline model ready! The final layer is now a standard linear layer:")
print(baseline_model.fc)

# 5. Load CIFAR-10 data using our helper function
train_loader, test_loader = get_cifar10_dataloaders(batch_size=128)

# 6. Setup Loss function and Optimizer
loss_fn = nn.CrossEntropyLoss()

# IMPORTANT: We ONLY pass the parameters of the final layer to the optimizer
optimizer = optim.Adam(baseline_model.fc.parameters(), lr=0.001)

# 7. Start the training! (Let's try 5 epochs to see if it works)
EPOCHS = 5
print(f"Starting training of Baseline ResNet-18 for {EPOCHS} epochs...")

baseline_results = train_model(
    model=baseline_model,
    train_loader=train_loader,
    test_loader=test_loader,
    optimizer=optimizer,
    loss_fn=loss_fn,
    epochs=EPOCHS,
    device=device,
)

# 8. Plot the loss and accuracy curves when training is finished
plot_loss_curves(baseline_results)
```
